```python
#Gradient Boosting
pipeline = Pipeline([
    ('gb', GradientBoostingRegressor(random_state=42))  # paso de modelo
])
# 2. Definir el grid de hiperparámetros
param_grid = {
    'gb__n_estimators': [500, 1000],
    'gb__learning_rate': [0.01, 0.05, 0.1],
    'gb__max_depth': [5, 10],
}
# 3. Definir GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=2
)
# 4. Entrenar
grid_search.fit(X_train, Y_train)
# 5. Mejor modelo
print("Mejores hiperparámetros:", grid_search.best_params_)
best_model_gb = grid_search.best_estimator_
# 6. Evaluar en test
Y_pred = best_model_gb.predict(X_test)
print("MSE:", mean_squared_error(Y_test, Y_pred))
print("R²:", r2_score(Y_test, Y_pred))
```

```python
#Support Vector Regressor
pipeline3 = Pipeline([('SVR1', SVR())])
# Definir la cuadrícula de hiperparámetros
param_grid3 = {
    'SVR1__kernel': ['rbf'],
    'SVR1__C': [5, 10, 100],
    'SVR1__epsilon': [0.01, 0.1, 0.5],
    'SVR1__gamma': [0.01, 0.1, 0.5],
}
# Crear GridSearchCV con 3 folds de validación cruzada
grid_search3 = GridSearchCV(
    estimator=pipeline3,
    param_grid=param_grid3,
    cv=3,
    scoring='r2',   # Métrica de evaluación
    n_jobs=-1
)
# Entrenar GridSearchCV
grid_search3.fit(X_train_std, Y_train)

# Mejor modelo
best_model_svr = grid_search3.best_estimator_
print("Mejores parámetros:", grid_search3.best_params_)

# Predicción sobre el set de prueba
Y_pred_svr = best_model_svr.predict(X_test_std)

# Evaluación
print("MSE:", mean_squared_error(Y_test, Y_pred_svr))
print("R²:", r2_score(Y_test, Y_pred_svr))
```

```python
#Modelo 1 Redes neuronales
# 1. Definir la función que crea el modelo
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

def create_model(neurons=64, activation='relu', optimizer='adam'):
    model = Sequential([
        Input(shape=(9,)),
        Dense(neurons, activation=activation),
        Dense(1, activation='linear')
    ])
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
# 2. Envolver con KerasRegressor
regressor = KerasRegressor(model=create_model, verbose=0, random_state=42)
# 3. Pipeline: red neuronal
pipeline = Pipeline([
    ('nn', regressor)
])

# 4. Definir la grilla de hiperparámetros
param_grid = {
    'nn__model__neurons': [32, 64, 128],      # número de neuronas en la capa oculta
    'nn__model__activation': ['relu'],
    'nn__model__optimizer': ['adam'],
    'nn__epochs': [200, 500],
    'nn__batch_size': [10, 20]
}
# 5. Grid Search
grid = GridSearchCV(estimator=pipeline,
                    param_grid=param_grid,
                    cv=3,
                    scoring=mse_scorer,
                    verbose=1)
# 6. Entrenar
grid_result = grid.fit(X_train_std, Y_train)
# 7. Resultados
print("Mejores parámetros:", grid_result.best_params_)
print("Mejor score (MSE negativo):", grid_result.best_score_)
# 8. Evaluar en test
best_model_rn = grid_result.best_estimator_
Y_pred_rn = best_model_rn.predict(X_test_std).flatten()
print("MSE:", mean_squared_error(Y_test, Y_pred_rn))
print("R²:", r2_score(Y_test, Y_pred_rn))
```
